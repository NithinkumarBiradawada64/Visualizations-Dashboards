```python
df["Length of Stay"]=pd.to_numeric(df["Length of Stay"],errors='coerce')
df["Birth Weight"]=pd.to_numeric(df["Birth Weight"],errors='coerce')
df["Total Charges"]=pd.to_numeric(df["Total Charges"],errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Hospital Service Area"].fillna("Unknow",inplace=True)
df["Hospital County"].fillna("Unknow",inplace=True)
df["Operating Certificate Number"].fillna(00000,inplace=True)
df["Permanent Facility Id"].fillna(0000,inplace=True)
df["Zip Code - 3 digits"].fillna("Unknow",inplace=True)
df["CCSR Diagnosis Code"].fillna("Unknow",inplace=True)
df["CCSR Diagnosis Description"].fillna("Unknow",inplace=True)
df["CCSR Procedure Code"].fillna("Unknow",inplace=True)
df["CCSR Procedure Description"].fillna("Unknow",inplace=True)
df["APR Severity of Illness Description"].fillna("Unknow",inplace=True)
df["APR Risk of Mortality"].fillna("Unknow", inplace=True)
df["Payment Typology 2"].fillna("Unknow", inplace=True)
df["Payment Typology 3"].fillna("Unknow", inplace=True)
df["Length of Stay"].fillna(0, inplace=True)
df["Birth Weight"].fillna(0, inplace=True)
```
